```python
# 모든 코스닥 데이터 하나로 합치기
kosdaq_stocks = pd.concat([kosdaq_19_stocks, kosdaq_20_stocks, kosdaq_21_stocks, kosdaq_22_stocks], ignore_index=True)

# 'Date' 컬럼 데이터 타입 변경 - object -> datetype
kosdaq_stocks['Date'] = pd.to_datetime(kosdaq_stocks['Date'])   # 'Date' 컬럼을 날짜 형식으로 변경
kosdaq_stocks['Date'] = kosdaq_stocks['Date'].dt.strftime('%Y-%m-%d')  # 'Date' 컬럼에서 초를 제거하고 'YYYY-MM-DD' 형식으로 변경

# 'KOSDAQ' 열에서 숫자가 아닌 값을 식별합니다.
non_numeric_values = kosdaq_stocks['KOSDAQ'].loc[~kosdaq_stocks['KOSDAQ'].astype(str).str.replace(',', '').str.isnumeric()]

# 숫자가 아닌 값을 처리하고 'KOSDAQ' 열을 부동 소수점으로 변환
kosdaq_stocks['KOSDAQ'] = kosdaq_stocks['KOSDAQ'].apply(
    lambda x: float(x.replace(',', '')) if isinstance(x, str) and x.replace(',', '').replace('.', '').isnumeric() else x
)

# 다른 열 데이터 유형 변경 object -> int
columns_to_convert = ['ECHOPRO_BM', 'HANA_METERIALS', 'SM_ENTER', 'JYP_ENTER', 'PERARLABYSS', 'SEOUL_SMC']
kosdaq_stocks[columns_to_convert] = kosdaq_stocks[columns_to_convert].apply(lambda x: x.str.replace(',', '').astype(int))

# 숫자가 아닌 값을 처리한 후 'KOSDAQ' 열을 부동 소수점으로 변환
kosdaq_stocks['KOSDAQ'] = pd.to_numeric(kosdaq_stocks['KOSDAQ'], errors='coerce')

kosdaq_stocks
```
