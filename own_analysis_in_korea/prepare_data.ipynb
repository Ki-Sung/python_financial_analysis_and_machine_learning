```python
# 날짜와 종가, 컬럼명 변경
kospi_21.rename(columns={"날짜":"Date", "종가":"KOSPI"}, inplace=True)
samsung_21.rename(columns={"날짜":"Date", "종가":"SAMSUNG"}, inplace=True)
hyundaicar_21.rename(columns={"날짜":"Date", "종가":"HYUNDAI_CAR"}, inplace=True)
kia_21.rename(columns={"날짜":"Date", "종가":"KIA"}, inplace=True)
kolmar_21.rename(columns={"날짜":"Date", "종가":"KORMAR"}, inplace=True)
naver_21.rename(columns={"날짜":"Date", "종가":"NAVER"}, inplace=True)
kakao_21.rename(columns={"날짜":"Date", "종가":"KAKAO"}, inplace=True)
hyundai_depat_21.rename(columns={"날짜":"Date", "종가":"THE_HYUNDAI"}, inplace=True)
```

```python
# 필요 없는 컬럼 제거
kospi_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
samsung_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
hyundaicar_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
kia_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
kolmar_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
naver_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
kakao_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
hyundai_depat_21.drop(columns=["시가", "고가", "저가", "거래량", "변동 %"], inplace=True)
```

```python
# 2021년 코스피 데이터 병합
dfs = [kospi_21, samsung_21, hyundaicar_21, kia_21, kolmar_21, naver_21, kakao_21, hyundai_depat_21]

kospi_21_stocks = pd.DataFrame({'Date': []})  # 빈 데이터프레임을 생성
for df in dfs:
    kospi_21_stocks = pd.merge(kospi_21_stocks, df, on='Date', how='outer')

# null 값 제거
kospi_21_stocks = kospi_21_stocks.dropna(axis=0)
kospi_21_stocks
```
